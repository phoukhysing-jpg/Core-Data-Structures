```python
coffee_shop = {
    "coffee_beans" : 50,
    "Cups" : 40
    
}
print("review")
print("sell")
print("restock")
print("exit")
coffee_beans = 50
cups = 40
while True:
    choice = input("Enter your choose").lower()
    if choice== 'exit':
        print("closing")
        break
    elif choice== 'review':
        print(coffee_shop)
    elif choice== 'sell':
        print("coffee_beans","Cups")
        option = input("Your choice").lower()
        amount = int(input("How many do you want?"))
        if option=='coffee_beans':
            coffee_beans = coffee_beans - amount
            print("Updated coffee_beans:",coffee_beans)
        elif  option=='cups' :
            cups =cups - amount
            print("Updated cups:",cups)
    elif choice == 'restock':
        print("coffee_beans","Cups")
        option = input("Your choice").lower()
        amount = int(input("How many do you want?"))
        if option=='coffee_beans' :
            coffee_beans = coffee_beans + amount
            print("Updated coffee_beans:",coffee_beans)
        elif option=='cups' :
            cups = cups + amount
            print("Updated cups:",cups)
    else:
        print("error")    
```
